**Gramáticas Livres-de-Contexto:**

Considere uma gramática livre-do-contexto descrita por G = <V, Σ, R, S>.
Sendo, V um
conjunto finito de variáveis, Σ um conjunto finito de terminadores disjunto de V, R um
conjunto finito de regras de produção e S ∈ V, é a variável inicial (start).
Implemente um programa que executa as regras de derivação da gramática. A configuração
da gramática pode ser feita de uma das seguintes maneiras:
* G é configurada no próprio código.
* A configuração de G é lida de um arquivo padronizado.
* A configuração de G é inserida via teclado.

Na fase de execução (derivação das regras), forneça a regra que será executada e mostre o
estado de derivação após a execução da regra selecionada. Para facilitar a implementação, numere as
regras e a cada interação apresente as regras que estão ativas, de tal modo que o usuário só pode
selecionar regras habilitadas (selecione a regra pelo seu número). Quando não houver mais regras
habilitadas, encerre o programa. Para melhorar a legibilidade, utilize as variáveis (elementos do
conjunto V) com letra em maiúsculo e os terminais (elementos de Σ) com dígitos ou letras em
minúsculo.
Teste a solução com ao menos duas gramáticas, como por exemplo a gramática para
reconhecer palíndromos ou a gramática G= <{A,B}, {0,1}, R, A>, com R dado por:
* A -> 0A1
* A -> B
* B -> # (vazio)

```python

# gramatica sugerida

'''
 A -> 0A1
 A -> B
 B -> #

'''
regra1 = [('A', '0A1'), ('A', 'B'), ('B', '#')]
g1 = '<{A,B}, {0,1}, R, A>'

#palindromo binario
'''
s -> 0s0
s -> 1s1
s -> 0
s -> 1
s -> #

'''
regra2 = [('S', '0S0'), ('S', '1S1'), ('S', '0'), ('S', '1'), ('S', '#')]
g2 = '<{S}, {0,1}, R, S>'


def derivacao_regra(regras, simbolo_inicial):

  print('\n---- Início da derivação ----\n')
  atual = simbolo_inicial

  while any(c.isupper() for c in atual):

    print('Variável atual:', atual, '\n')
    print('Regras disponíveis: \n')
    regras_aplicaveis = []

    for i  in range(len(regras)):
      variavel, regra = regras[i]
      if variavel in atual:
        regras_aplicaveis.append((i, variavel, regra))
        print(f'[{i}] -> {regras[i]}')

    escolha = int(input('\nEscolha uma regra: '))
    encontrada = False

    for i, variavel, regra in regras_aplicaveis:
      if escolha == i:
        atual = atual.replace(variavel, regra, 1)
        print('\nResultado:', atual)
        encontrada = True
        break

    if not encontrada:
      print('Regra inválida.')

  if '#' in atual:
    atual = atual.replace('#', '')


  print("\nCadeia gerada", atual)


print("Escolha uma gramática:")
print("[1] Gramática com a regra 1")
print("[2] Gramática com a regra 2")

opcao = input("Opção: ")

if opcao == '1':
    print('\nGramática escolhida:', g1, 'e regra da gramática(R):', regra1)
    derivacao_regra(regra1, 'A')

elif opcao == '2':
    print('\nGramática escolhida:', g2, 'e regra da gramática(R):', regra2)
    derivacao_regra(regra2, 'S')

else:
    print("Opção inválida")
    exit()
```
